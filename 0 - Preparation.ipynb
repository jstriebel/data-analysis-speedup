```python
from pathlib import Path

from utils.preparation import (
    get_chunks,
    get_segmentation,
    get_stats,
    plot,
    save_chunked_stats,
)

segmentation = get_segmentation()
plot(segmentation)
chunked_segmentation = list(get_chunks(segmentation, chunks=3))
plot(chunked_segmentation)
chunked_stats = get_stats(chunked_segmentation)
plot(chunked_segmentation, chunked_stats[:-1])
stats = get_stats(segmentation)
plot(segmentation, stats[:-1])

save_chunked_stats(chunked_stats, Path("data/small"))
```

```python
segmentation = get_segmentation(size=1000, segments=50, radius=0.05)
plot(segmentation)
chunked_segmentation = list(get_chunks(segmentation, chunks=10))
chunked_stats = get_stats(chunked_segmentation, repeat=10, progress=True)

save_chunked_stats(chunked_stats, Path("data/medium"), progress=True)
```
